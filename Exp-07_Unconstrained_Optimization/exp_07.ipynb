```python
import numpy as np
from scipy.optimize import least_squares

# -------------------------------
# Model function (logistic growth)
def model(x, t):
    a, b, c = x
    return a / (1 + b * np.exp(-c * t))

# Residual function
def residual(x, t, y):
    return model(x, t) - y

# -------------------------------
# Generate noisy data
np.random.seed(42)
t = np.linspace(0, 5, 60)

# True parameters
true_params = [10, 5, 1.2]

# Generate observations
y = model(true_params, t) + 0.5 * np.random.randn(len(t))

# -------------------------------
# Initial guess
x0 = [8, 3, 1]

# Fit model
result = least_squares(residual, x0, args=(t, y))

# -------------------------------
# Output
print("Optimized parameters:", result.x)
```
