```python
import numpy as np

def stochastic_gradient_descent(X, y, learning_rate=0.02, epochs=800):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)

    for epoch in range(epochs):
        for i in range(n_samples):
            prediction = np.dot(X[i], weights)
            error = prediction - y[i]
            gradient = error * X[i]
            weights -= learning_rate * gradient

    return weights

# -------------------------------
# Example usage
if __name__ == "__main__":
    
    # Generate sample data
    np.random.seed(42)
    X = 3 * np.random.rand(80, 1)
    y = 2 + 5 * X + np.random.randn(80, 1)

    # Add bias term (x0 = 1)
    X_b = np.c_[np.ones((80, 1)), X]

    # Flatten y
    y = y.ravel()

    # Parameters
    learning_rate = 0.02
    epochs = 800

    # Run SGD
    weights = stochastic_gradient_descent(X_b, y, learning_rate, epochs)

    print("Learned Weights:", weights)
```
